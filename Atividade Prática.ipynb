### <span style="color:red"> A. Usando a geração de números aleatórios em python, conforme aula anterior, avalie graficamente e numericamente a distribuição de probabilidade da média amostral $\overline{X}$. </span>
<br />
Você vai precisar de uma amostra de médias. Portanto, siga o seguinte roteiro: <br /> 

### 1. Gere um conjunto de 50 amostras aleatórias de tamanhos iguais a $n$;<br />

```python
np.random.seed(seed=0)

X = np.random.random(50)
print(X)
```

### 2. Calcule as médias das 50 amostras geradas;<br />

```python
media = np.mean(X)
display(media)
```

### 3. Crie o histograma para a amostra de 50 médias resultantes;<br />

```python
hist, bin_edges = np.histogram(X, bins=50)
```

### 4. Plote a curva da distribuição sobre o histograma;<br />

```python
fig, ax = plt.subplots()
ax.hist(X, bin_edges, cumulative=False, density=True)
mu = 0.14055927231309787
sigma = 1
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 50)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
ax.set_xlabel('x')
ax.set_ylabel('Frequencia')
plt.show()
```

### 5. Realize o teste de aderência e avalie a normalidade.<br />

```python
stats.shapiro(X)
```

```python
np.random.seed(seed=0)
xn = np.random.random(10)
peso = xn * 10 + 70;
hist, bin_edges = np.histogram(peso, bins=10)
fig, ax = plt.subplots()
ax.hist(peso, bin_edges, cumulative=False, density=True, color='blue')
#sigma = math.sqrt(s2)
#média = 70 e desvio-padrão = 10
mu = 70
sigma = 10
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
ax.set_xlabel('Peso')
ax.set_ylabel('Frequencia')
plt.show()
```

### <span style="color:red">B. Realize as etapas de 1 a 5 acima para $n \in {5, 10, 20, 30, 50, 100}$. Discuta os resultados em função dessa variação de $n$. <br /> </span>

## RANDOM

## $n \in {5}$

```python
np.random.seed(seed=0)

lista_media = []
m = 0
for y in range(50):
    X = np.random.random(5)
    print(f"""{y+1} {X} 
    
          Média = {np.mean(X)}
                              
          """)
    m += np.mean(X)
    lista_media.append(np.mean(X))
    
print(f'Média total {m}')
```

```python
hist, bin_edges = np.histogram(lista_media,bins=7)
fig, ax = plt.subplots()
ax.hist(lista_media, bin_edges, cumulative=False, density=True, color='blue')
mu = np.mean(lista_media)
sigma = np.std(lista_media)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
ax.set_xlabel('x')
ax.set_ylabel('Frequencia')
plt.show()
```

```python
stats.shapiro(lista_media)
```
